# pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc=’mean’)

```python
#data : DataFrame
#values : column to aggregate, optional
#index: column, Grouper, array, or list of the previous
#columns: column, Grouper, array, or list of the previous
#aggfunc: function, list of functions, dict, default numpy.mean
```

```python
import pandas as pd
from pandas import DataFrame
df1 = DataFrame(
                [
                    ['Aryan','Singh','Full-time','adminstration',2,20000],
                    ['Rohan','Agarawal','intern','Technical',3,5000],
                    ['Rjya','Shah','Full-time','adminstration',5,10000],
                    ['Yash','Bhatia','Part-time','Technical',7,10000],
                    ['Siddhant','Khanna','Full-time','Management',6,20000],
                ], columns=['First Name','Last name','Type','Department','YoE','Salary']
              )
df1
```

```python
table = pd.pivot_table(df1,index ="Type")
table
```

```python
table = pd.pivot_table(df1,index ="Type", values="Salary")
  
table
```

```python
p_table = pd.pivot_table(df1,index="Type", columns="Department",values="Salary")
p_table
```

```python
p_table = pd.pivot_table(df1,index="Type", columns="Department",values="Salary",aggfunc='sum')
p_table
```
